```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel(r'C:\Users\Samsung\Desktop\Dados_Empresas_Energia.xlsx')
```

```python
dados.head()
```

```python
dados.info()
```

```python
#transformar a coluna de data no index
dados.set_index('Data', inplace=True)
```

```python
dados.head()
```

```python
#plotar grafico de linha
dados.plot(figsize=(15,6))
```

```python
#melhorar gráfico

#estilo do grafico
plt.style.use('seaborn-darkgrid')

#tamanho do grafico
plt.figure(figsize=(16,7))

#titulo
plt.title('Ações de empresas de energia')

#plot da petrobras
plt.plot(dados['Petrobras'], label='Petrobras', color = '#008c4a', linewidth=4)
plt.text(dados.index[-1], dados['Petrobras'].tail(1), 'Petrobras', color = '#008c4a', fontsize=14)

# Plot de todas as colunas
for Coluna in dados.columns[1:]:
  # Plot das outras ações
  plt.plot( dados.index, dados[Coluna], color='gray', linewidth=2, alpha=0.7 )
  # Texto das outras ações
  plt.text( dados.index[-1], dados[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left' )


#eixo x
plt.xlabel('Período')
#eixo y
plt.ylabel('Preço de Fechamento');
```
